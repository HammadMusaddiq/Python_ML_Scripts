```python
from openpyxl import load_workbook
import numpy as np
import pandas as pd
import csv
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
```

```python
list0=[]
wb = load_workbook(filename='G1_2018-12-04.xlsx', read_only=True)
ws = wb['Sheet']
for row in ws.rows:
    list00=[]
    for kl in range(len(row)):
        
        list00.append(row[kl].value)
    list0.append(list00)
```

```python
data_watch_p=pd.DataFrame(list0[1:], columns=['date','time','distance','enhanced_speed','speed','heart_rate','latitude','longitude'])
data_watch_p.head(10)
```

```python
data_watch_p['time']
```

```python
data_watch=np.array(data_watch_p.iloc[:,:])
```

```python
txt=open('Pha_coordinates.kml','r')
datatxt=txt.read()
datatxt=datatxt.split('coordinates')[-2]
datatxt=datatxt.replace(' ','').split('\n')[1:-1]

Datalatlon=[]
for i in range(len(datatxt)):
    Datalatlon.append([float(datatxt[i].split(',')[-2]),float(datatxt[i].split(',')[-3])])

Datalatlon=np.array(Datalatlon)
lons_lats_vect = np.column_stack((Datalatlon[:,1], Datalatlon[:,0])) # Reshape coordinates
lons_lats_vect
```

```python
polygon = Polygon(lons_lats_vect) # create polygon
polygon
```

```python
point = Point(float(data_watch[-1,-1]),float(data_watch[-1,-2])) #(position_long,position_lat)
print(polygon.contains(point)) # check if polygon contains point
point
```

```python
geoJsonData = {
    "features": [
        {
            "geometry": {
                "coordinates": [
                    [
                        lons_lats_vect[0][0],
                        lons_lats_vect[0][1]
                    ],
                    [
                        lons_lats_vect[1][0],
                        lons_lats_vect[1][1]
                    ],
                    [
                        lons_lats_vect[2][0],
                        lons_lats_vect[2][1]
                    ],
                    [
                        lons_lats_vect[3][0],
                        lons_lats_vect[3][1]
                    ],
                    [
                        lons_lats_vect[4][0],
                        lons_lats_vect[4][1]
                    ]
                ],
                "type": "LineString"
            },
            "properties": {
                "stroke": "#fc1717",
                "stroke-opacity": 1,
                "stroke-width": 3
            },
            "type": "Feature"
        },
#         {
#             "geometry": {
#                 "coordinates": [
#                     [
#                         lons_lats_vect[0][0]+0.005,
#                         lons_lats_vect[0][1]+0.005
#                     ],
#                     [
#                         lons_lats_vect[1][0]+0.005,
#                         lons_lats_vect[1][1]+0.005
#                     ],
#                     [
#                         lons_lats_vect[2][0]+0.005,
#                         lons_lats_vect[2][1]+0.005
#                     ],
#                     [
#                         lons_lats_vect[3][0]+0.005,
#                         lons_lats_vect[3][1]+0.005
#                     ],
#                     [
#                         lons_lats_vect[4][0]+0.005,
#                         lons_lats_vect[4][1]+0.005
#                     ]
#                 ],
#                 "type": "LineString"
#             },
#             "properties": {
#                 "stroke": "#1f1a95",
#                 "stroke-opacity": 1,
#                 "stroke-width": 2
#             },
#             "type": "Feature"
#         }
    ],
    "type": "FeatureCollection"
}
```

```python
import folium
m = folium.Map(location=[ lons_lats_vect[4][1], lons_lats_vect[4][0]], zoom_start=15)
folium.GeoJson(geoJsonData,
    style_function=lambda x: {
        'color' : x['properties']['stroke'],
        'weight' : x['properties']['stroke-width'],
        'opacity': 0.6,
        }).add_to(m)
m
```

```python
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()
```

```python
tok=None
print('Activity Starting Time:',data_watch[0,1])
for i in range(0,len(data_watch)):
    point = Point(float(data_watch[i,7]),float(data_watch[i,6])) #(position_long,position_lat)
    if polygon.contains(point):
        if tok!=0:
            print ('Entering Time:',data_watch[i,1])
            tok=0
            Entering_Time_index=i
            
        incidents.add_child(
            folium.features.CircleMarker(
                [float(data_watch[i,6]), float(data_watch[i,7])],
                radius=3, # define how big you want the circle markers to be
                color='yellow',
                fill=True,
                fill_color='blue',
                fill_opacity=0.6
            )
        )
    else:
        if tok!=1:
            print ('Outside Time:',data_watch[i,1])
            tok=1
            Outside_Time_index=i
        
        incidents.add_child(
            folium.features.CircleMarker(
                [float(data_watch[i,6]), float(data_watch[i,7])],
                radius=3, # define how big you want the circle markers to be
                color='green',
                fill=True,
                fill_color='blue',
                fill_opacity=0.6
            )
        )
print('Activity End Time:',data_watch[-1,1])
```

```python
m.add_child(incidents)
```

```python
from folium import plugins
import folium
m = folium.Map(location=[ lons_lats_vect[4][1], lons_lats_vect[4][0]], zoom_start=15)
folium.GeoJson(geoJsonData,
    style_function=lambda x: {
        'color' : x['properties']['stroke'],
        'weight' : x['properties']['stroke-width'],
        'opacity': 0.6,
        }).add_to(m)


# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(m)

# loop through the dataframe and add each data point to the mark cluster
for i in range(0,len(data_watch)):
    folium.Marker(
        location=[float(data_watch[i,6]), float(data_watch[i,7])],
        icon=None,
    ).add_to(incidents)

# display map
m
```
